````python
def value_to_symbol(value):
    if value == 0:
        return '  '
    elif value == 1:
        return 'X '
    elif value == 2:
        return 'O '

def draw_board(input_board):
    n = len(input_board)
    m = len(input_board[0])
    horizontal_line = '+' + '---+' * m
    for row in range(n):
        print(horizontal_line)
        row_text = '|'
        for col in range(m):
            row_text += ' ' + value_to_symbol(input_board[row][col]) + '|'
        print(row_text)
    print(horizontal_line)
````

*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
def check_game_state(test_game_board):
    # The dimension of our game board.
    list_dim = len(test_game_board[0])

    # Initializing the game state variables.
    winner = None # The win condition in each test will return the winner with this variable.

    # Checking for column-wise win condition.
    element_list = [item for row in test_game_board for item in row] # Nested list comprehension to flatten the list.
    # We will iterate over the columns and check if the set is of length one and not zero.
    for m in range(list_dim):
        if len(set(element_list[m::list_dim])) == 1 and element_list[m] != 0:
            return element_list[m]

    # Checking for diagonal win condition.
    diagonal_elements = []
    for element in range(list_dim):
        diagonal_elements.append(test_game_board[element][element])
    if len(set(diagonal_elements)) == 1 and diagonal_elements[0] != 0:
        return diagonal_elements[0]

    # Checking for anti-diagonal win condition.
    anti_diagonal_elements = []
    for element in range(list_dim):
        anti_diagonal_elements.append(test_game_board[element][list_dim - 1 - element])
    if len(set(anti_diagonal_elements)) == 1 and anti_diagonal_elements[0] != 0:
        return anti_diagonal_elements[0]

    # Check row-wise win condition.
    for row in test_game_board:
        if len(set(row)) == 1 and row[0] != 0:
            return row[0]

    # A game is a draw if the game board has no empty spaces and no winner.
    if 0 not in element_list: # If 0 is not in element_list, game board is full.
        return 0

    # Game incomplete condition.
    if winner is None:
        return -1
````

````python
winner_is_2 = [[2, 2, 0],
               [2, 1, 0],
               [2, 1, 1]]

winner_is_1 = [[1, 2, 0],
               [2, 1, 0],
               [2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
                    [2, 1, 0],
                    [2, 1, 1]]

no_winner = [[1, 2, 0],
             [2, 1, 0],
             [2, 1, 2]]

also_no_winner = [[1, 2, 0],
                  [2, 1, 0],
                  [2, 1, 0]]

# Additional game states for testing.
row_winner_1 = [[0, 2, 2],
                [0, 1, 0],
                [1, 1, 1]]

row_winner_2 = [[2, 2, 2, 0],
                [0, 1, 0, 2],
                [2, 2, 2, 2],
                [1, 2, 0, 2]]

draw = [[2, 1, 2],
        [2, 1, 2],
        [1, 2, 1]]
````

*Exercise 5:* Write a function that takes a game board, player number, and `(x,y)` coordinates and places "X" or "O" in the correct location of the game board. Make sure that you only allow filling previously empty locations. Return `True` or `False` to indicate successful placement of "X" or "O".

````python
# Initializing the starting game board
ex_5_board = create_board(3)

def modify_game_board(game_board, player_number, coords):
    row, col = coords
    n = len(game_board)
    m = len(game_board[0])
    if not (0 <= row < n and 0 <= col < m):
        return False
    if game_board[row][col] == 0:
        game_board[row][col] = player_number
        return True
    else:
        return False
````

*Exercise 6:* Modify Exercise 3 to show column and row labels so that players can specify location using "A2" or "C1".

````python
def draw_board(input_board):
    n = len(input_board)
    m = len(input_board[0])
    horizontal_line = '  +' + '---+' * m
    header = '  ' + ' '.join(f'{c + 1:^3}' for c in range(m))
    print(header)
    for row in range(n):
        print(horizontal_line)
        label = chr(ord('A') + row)
        row_text = f'{label} |'
        for col in range(m):
            row_text += ' ' + value_to_symbol(input_board[row][col]) + '|'
        print(row_text)
    print(horizontal_line)
````

*Exercise 7:* Write a function that takes a board, player number, and location specified as in exercise 6 and then calls exercise 5 to correctly modify the board.  

````python
def place_by_label(game_board, player_number, location):
    letter = location[0].upper()
    number = location[1:]

    row = ord(letter) - ord('A')
    col = int(number) - 1

    return modify_game_board(game_board, player_number, (row, col))
````

*Exercise 8:* Write a function is called with a board and player number, takes input from the player using python's `input`, and modifies the board using your function from exercise 7. Note that you should keep asking for input until you have gotten a valid input that results in a valid move.

````python
def get_player_move(game_board, player_number):
    while True:
        location = input(f'Player {player_number}, enter your move (e.g. A2): ').strip().upper()

        # Validate format.
        if len(location) < 2 or not location[0].isalpha() or not location[1:].isdigit():
            print('Invalid format. Must follow this format: A2 for example.')
            continue

        if place_by_label(game_board, player_number, location):
            return location
        else:
            print('Invalid move. Try again.')
````

*Exercise 9:* Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

````python
def play_game(board_size=3):
    game_board = create_board(board_size)
    current_player = 1

    while True:
        draw_board(game_board)

        get_player_move(game_board, current_player)

        state = check_game_state(game_board)
        if state != -1:
            draw_board(game_board)
            if state == 0:
                print('The game is a draw.')
            else:
                print(f'Player {state} wins!')
            break
        current_player = 2 if current_player == 1 else 1
````
