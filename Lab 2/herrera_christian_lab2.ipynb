*Exercise 3:* Modify exercise 2, so that it takes a matrix of the form from exercise 1 and draws a tic-tac-tie board with "X"s and "O"s.  

````python
def value_to_symbol(value):
    if value == 0:
        return '  '
    elif value == 1:
        return 'X '
    elif value == 2:
        return 'O '

def draw_board(input_board):
    n = len(input_board)
    m = len(input_board[0])
    horizontal_line = '+' + '---+' * n
    for row in range(n):
        print(horizontal_line)
        row_text = '|'
        for col in range(m):
            row_text += ' ' + value_to_symbol(input_board[row][col]) + '|'
        print(row_text)
    print(horizontal_line)
````

````python
board = create_board(3)
board[0][0] = 1
board[0][2] = 2
board[1][0] = 1
draw_board(board)
````
