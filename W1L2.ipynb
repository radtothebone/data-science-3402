1. Game Representation

```python
# Index assignment for matrix representation of the game board
player_1 = 1
player_2 = 2
empty = 0

size = 8

# Create the board first, put the pieces on top

board = list()

for i in range(size):
    row = list()
    for j in range(size):
        row.append(empty)
    board.append(row)

board
```

```python
# Much more compact definition of the board
board = [[empty]*size]*size
board
```

```python
# Undesired behavior
board[1][1] = 1
board
```

```python
board = list()

board = [[empty]*size for _ in range(size)]

board
```

```python
def make_game_board(size):
    # Make the empty board
    board = list()
    board = [[empty]*size for _ in range(size)]
    # Column first approach
    for i in range(0,size,2):
        board[1][i] = player_1
        board[-1][i] = player_2
        board[-3][i] = player_2
    for i in range(1,size,2):
        board[0][i] = player_1
        board[2][i] = player_1
        board[-2][i] = player_2
    return board

make_game_board(8)
```
