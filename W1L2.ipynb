```python
# Index assignment for matrix representation of the game board
player_1 = 1
player_2 = 2
empty = 0

size = 8

# Create the board first, put the pieces on top

board = list()

for i in range(size):
    row = list()
    for j in range(size):
        row.append(empty)
    board.append(row)

board
```

```python
# Much more compact definition of the board
board = [[empty]*size]*size
board
```

```python
def make_game_board(size):
    # Make the empty board
    board = list()
    board = [[empty]*size for _ in range(size)]

    for i in range(0,size,2):
        board[1][i] = player_1
        board[-1][i] = player_2
        board[-3][i] = player_2

    for i in range(1,size,2):
        board[0][i] = player_1
        board[2][i] = player_1
        board[-2][i] = player_2

    return board

make_game_board(8)
```

# Game Rules
### Rules for movement:
Regular pieces can only move in the forward direction, offset to the left or right by one. 

```python
left_move = 0
right_move = 1

def player_moves(player, direction):
    if player == player_1:
        if direction == left_move:
            return (1,1)
        elif direction == right_move:
            return (1, -1)
        
    if player == player_2:
        if direction == left_move:
            return (-1,-1)
        elif direction == right_move:
            return (-1, 1)
        
player_1_left_move = (1,1)
player_1_right_move = (1,-1)

player_2_left_move = (-1,-1)
player_2_right_move = (-1,1)

moves = {player_1:{left_move:player_1_left_move,
                   right_move:player_1_right_move},
         player_2:{left_move:player_2_left_move,
                   right_move:player_2_right_move}}



moves[player_1][right_move]
```

```python
def print_message(message, verbose=True):
    if verbose:
        print(message)

def move_piece(board, player, location, move, verbose=True):
    x, y = location

    # Check if the players piece is at correct location
    if not board[x][y] == player:
        print_message('Player does not have piece at correct location', verbose=True)
        return False
    
    x_offset, y_offset = moves[player][move]

    move_possible = x + x_offset and x + x_offset >= 0 and y + y_offset < size and y + y_offset >= 0

    jump_possible = x + 2*x_offset and x + 2*x_offset >= 0 and y + 2*y_offset < size and y + 2*y_offset >= 0

    if not (move_possible or jump_possible):
        print_message("Move is out of the board", verbose=True)
        return False
    
    if move_possible and board[x + x_offset][y + y_offset] == empty:
        board[x][y] = empty
        board[x + x_offset][y + y_offset] = player
        print_message("Moved", verbose=True)
        return True
    elif jump_possible and board[x + x_offset][y + y_offset] != player and board[x + 2*x_offset][y + 2*y_offset] == empty:
        board[x][y] = empty
        board[x + x_offset][y + y_offset] = empty
        board[x + 2*x_offset][y + 2 * y_offset] = player

    else:
        return False
```
