```python
player_1 = 'B'
player_2 = 'W'
empty = 0

size = 8

def make_coord_list(x_offset, board_size):
    return [[0, x_offset],
            [0, (board_size - 1) - x_offset],
            [(board_size - 1), x_offset],
            [(board_size - 1), (board_size - 1) - x_offset]]

def make_game_board(board_size: int):
    board = [[empty] * board_size for _ in range(board_size)]

    for i in range(0, board_size):
        board[0][i] = player_1
        board[1][i] = player_1

        board[-1][i] = player_2
        board[-2][i] = player_2

    # Adding the pawn pieces across the rows
    for i in range(0, board_size):
        board[1][i] += 'P'
        board[board_size - 2][i] += 'P'

    # Adding the Rook pieces
    for coord in make_coord_list(0, board_size):
        board[coord[0]][coord[1]] += 'R'

    # Adding the Knight pieces
    for coord in make_coord_list(1, board_size):
        board[coord[0]][coord[1]] += 'N'

    # Adding the Bishop pieces
    for coord in make_coord_list(2, board_size):
        board[coord[0]][coord[1]] += 'B'

    # Adding the King pieces.
    k_coords = [[0,4], [board_size - 1,4]]
    for coord in k_coords:
        board[coord[0]][coord[1]] += 'K'

    # Adding the Queen pieces.
    q_coords = [[0,3], [board_size - 1, 3]]
    for coord in q_coords:
        board[coord[0]][coord[1]] += 'Q'

    return board

make_game_board(size)
```
