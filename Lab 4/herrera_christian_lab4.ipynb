1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, count, increment=1, max_value=None):
        self.count = count
        self.increment = increment
        self.max_value = max_value

    def increment_count(self):
        increment_val = self.count + self.increment
        if increment_val > self.max_value:
            print('Count is too high! Initialize reset immediately.')
        else:
            self.count += self.increment
            return self.count

    def reset_count(self):
        self.count = 0

    def display(self):
        print(self.count)
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:
    def __init__(self, count=0, increment=1, max_value=None):
        self.__count = count
        self.__increment = increment
        self.__max_value = max_value

    def increment_count(self):
        next_val = self.__count + self.__increment
        if next_val > self.__max_value:
            print('Error: Count is too high!')
        else:
            self.__count += self.__increment

    def reset_count(self):
        self.__count = 0

    def get_count(self):
        return self.__count

    def get_max_value(self):
        return self.__max_value

    def is_at_max(self):
        return self.__count == self.__max_value
```

```python
my_counter = Counter(count=0, increment=1, max_value=3)

my_counter.increment_count()
my_counter.increment_count()
my_counter.increment_count()

print('Current count:', my_counter.get_count())
print('Max value:', my_counter.get_max_value())
print('At max?', my_counter.is_at_max())

my_counter.increment_count()

my_counter.reset_count()
print('After reset:', my_counter.get_count())
print('At max?', my_counter.is_at_max())
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
# Circle: evenly spaced, 16 total
# Rectangle: 4 corners + 3 each side, 16 total
# Triangle: 3 corners + 4 each side, 15 total
import math # Needed for the trig.

class Shape:
	def calculate_area(self):
		raise NotImplementedError

	def calculate_perimeter(self):
		raise NotImplementedError

	def generate_coords(self):
		raise NotImplementedError
```

```python
class Circle(Shape):
	def __init__(self, radius, x_coord, y_coord):
		self.__radius = radius
		self.__x_coord = x_coord
		self.__y_coord = y_coord

	def calculate_area(self):
		return 3.14159 * self.__radius ** 2

	def calculate_perimeter(self):
		return 2 * 3.14159 * self.__radius

	def get_radius(self):
		return self.__radius

	def get_x_coord(self):
		return self.__x_coord

	def get_y_coord(self):
		return self.__y_coord

	def generate_coords(self):
		# Polar coords make this problem much easier.
		theta_vals = [i*math.pi/8 for i in range(16)]
		x_y_coords = [(round(self.__radius * math.cos(i) + self.__x_coord, 2),
					   round(self.__radius * math.sin(i) + self.__y_coord, 2)) for i in theta_vals]
		return x_y_coords

my_circle = Circle(radius=1, x_coord=0, y_coord=0)
my_circle.calculate_perimeter()
my_circle.generate_coords()
```

```python
class Rectangle(Shape):
	def __init__(self, length, width, x_coord, y_coord):
		self.__length = length
		self.__width = width
		self.__x_coord = x_coord
		self.__y_coord = y_coord

	def calculate_area(self):
		return self.__length * self.__width

	def calculate_perimeter(self):
		return 2 * self.__length + 2 * self.__width

	def get_length(self):
		return self.__length

	def get_width(self):
		return self.__width

	def get_x(self):
		return self.__x_coord

	def get_y(self):
		return self.__y_coord

	def generate_coords(self):
		points = []
		# bottom side (left to right)
		for i in range(4):
			points.append((round(self.__x_coord + self.__length * i/4, 2), round(self.__y_coord, 2)))
		# right side (bottom to top)
		for i in range(4):
			points.append((round(self.__x_coord + self.__length, 2), round(self.__y_coord + self.__width * i/4, 2)))
		# top side (right to left)
		for i in range(4):
			points.append((round(self.__x_coord + self.__length * (1 - i/4), 2), round(self.__y_coord + self.__width, 2)))
		# left side (top to bottom)
		for i in range(4):
			points.append((round(self.__x_coord, 2), round(self.__y_coord + self.__width * (1 - i/4), 2)))
		return points

my_rectangle = Rectangle(length=4, width=2, x_coord=0, y_coord=0)
my_rectangle.generate_coords()
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
class Shape:
	def calculate_area(self):
		raise NotImplementedError

	def calculate_perimeter(self):
		raise NotImplementedError

	def generate_coords(self):
		raise NotImplementedError

	def coord_check(self, x, y):
		raise NotImplementedError

class Rectangle(Shape):
	def __init__(self, length, width, x_coord, y_coord):
		self.__length = length
		self.__width = width
		self.__x_coord = x_coord
		self.__y_coord = y_coord

	def calculate_area(self):
		return self.__length * self.__width

	def calculate_perimeter(self):
		return 2 * self.__length + 2 * self.__width

	def get_length(self):
		return self.__length

	def get_width(self):
		return self.__width

	def get_x(self):
		return self.__x_coord

	def get_y(self):
		return self.__y_coord

	def generate_coords(self):
		points = []
		# bottom side (left to right)
		for i in range(4):
			points.append((round(self.__x_coord + self.__length * i/4, 2), round(self.__y_coord, 2)))
		# right side (bottom to top)
		for i in range(4):
			points.append((round(self.__x_coord + self.__length, 2), round(self.__y_coord + self.__width * i/4, 2)))
		# top side (right to left)
		for i in range(4):
			points.append((round(self.__x_coord + self.__length * (1 - i/4), 2), round(self.__y_coord + self.__width, 2)))
		# left side (top to bottom)
		for i in range(4):
			points.append((round(self.__x_coord, 2), round(self.__y_coord + self.__width * (1 - i/4), 2)))
		return points

	def coord_check(self, x, y):
		return self.__x_coord <= x <= self.__x_coord + self.__length and \
		       self.__y_coord <= y <= self.__y_coord + self.__width

my_rectangle = Rectangle(length=4, width=3, x_coord=0, y_coord=0)
print(my_rectangle.coord_check(2, 1))
print(my_rectangle.coord_check(5, 1))
print(my_rectangle.coord_check(4, 3))
```

```python
class Circle(Shape):
	def __init__(self, radius, x_coord, y_coord):
		self.__radius = radius
		self.__x_coord = x_coord
		self.__y_coord = y_coord

	def calculate_area(self):
		return 3.14159 * self.__radius ** 2

	def calculate_perimeter(self):
		return 2 * 3.14159 * self.__radius

	def get_radius(self):
		return self.__radius

	def get_x_coord(self):
		return self.__x_coord

	def get_y_coord(self):
		return self.__y_coord

	def generate_coords(self):
		theta_vals = [i*math.pi/8 for i in range(16)]
		x_y_coords = [(round(self.__radius * math.cos(i) + self.__x_coord, 2),
					   round(self.__radius * math.sin(i) + self.__y_coord, 2)) for i in theta_vals]
		return x_y_coords

	def coord_check(self, x, y):
		return (x - self.__x_coord) ** 2 + (y - self.__y_coord) ** 2 <= self.__radius ** 2

my_circle = Circle(radius=5, x_coord=0, y_coord=0)
print(my_circle.coord_check(0, 0))
print(my_circle.coord_check(3, 4))
print(my_circle.coord_check(6, 0))
```

```python
class Triangle(Shape):
	def __init__(self, leg_1, leg_2, x_coord, y_coord):
		self.__leg_1 = leg_1
		self.__leg_2 = leg_2
		self.__hyp = (leg_1 ** 2 + leg_2 ** 2) ** 0.5
		self.__x_coord = x_coord
		self.__y_coord = y_coord

	def calculate_area(self):
		return 0.5 * self.__leg_1 * self.__leg_2

	def calculate_perimeter(self):
		return self.__leg_1 + self.__leg_2 + self.__hyp

	def get_leg_1(self):
		return self.__leg_1

	def get_leg_2(self):
		return self.__leg_2

	def get_hyp(self):
		return self.__hyp

	def get_x(self):
		return self.__x_coord

	def get_y(self):
		return self.__y_coord

	def generate_coords(self):
		points = []
		for i in range(5):
			points.append((round(self.__x_coord + self.__leg_1 * i/5, 2), round(self.__y_coord, 2)))
		for i in range(5):
			points.append((round(self.__x_coord + self.__leg_1 * (1 - i/5), 2), round(self.__y_coord + self.__leg_2 * i/5, 2)))
		for i in range(5):
			points.append((round(self.__x_coord, 2), round(self.__y_coord + self.__leg_2 * (1 - i/5), 2)))
		return points

	def coord_check(self, x, y):
		in_bounding_box = self.__x_coord <= x <= self.__x_coord + self.__leg_1 and \
		                  self.__y_coord <= y <= self.__y_coord + self.__leg_2
		below_hyp = (x - self.__x_coord) / self.__leg_1 + (y - self.__y_coord) / self.__leg_2 <= 1
		return in_bounding_box and below_hyp

my_triangle = Triangle(leg_1=4, leg_2=4, x_coord=0, y_coord=0)
print(my_triangle.coord_check(1, 1))    # inside
print(my_triangle.coord_check(3, 3))    # above hypotenuse
print(my_triangle.coord_check(5, 1))    # outside bounding box
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
def overlaps(self, other):
	for x, y in self.generate_coords():
		if other.coord_check(x, y):
			return True
	for x, y in other.generate_coords():
		if self.coord_check(x, y):
			return True
	return False


# Monkey patch because redefining the classes over and over is making this notebook very long.
Shape.overlaps = overlaps


my_circle = Circle(radius=5, x_coord=0, y_coord=0)
close_circle = Circle(radius=5, x_coord=3, y_coord=4)
far_circle = Circle(radius=2, x_coord=20, y_coord=20)
print(my_circle.overlaps(close_circle)) # True
print(my_circle.overlaps(far_circle)) # False, far apart

my_rectangle = Rectangle(length=4, width=3, x_coord=0, y_coord=0)
overlapping_rectangle = Rectangle(length=4, width=3, x_coord=2, y_coord=1)
far_rectangle = Rectangle(length=2, width=2, x_coord=10, y_coord=10)
print(my_rectangle.overlaps(overlapping_rectangle))  # True, corners overlap
print(my_rectangle.overlaps(far_rectangle)) # False, far apart agian

my_triangle = Triangle(leg_1=4, leg_2=4, x_coord=0, y_coord=0)
far_triangle = Triangle(leg_1=2, leg_2=2, x_coord=10, y_coord=10)
print(my_triangle.overlaps(my_rectangle)) # True
print(my_triangle.overlaps(far_triangle)) # False: far apart

overlapping_circle = Circle(radius=3, x_coord=2, y_coord=1)
print(my_rectangle.overlaps(overlapping_circle)) # mixed shapes, true
```
