### Loop pattern

```python
for index in range(10):
    print(index)
```

```python
char = ['a', 'b', 'c']
for index, item in enumerate(char):
    print(index, item)
```

```python
char = ['a', 'b', 'c']
cap_char = ['A', 'B', 'C']

for item_1, item_2 in zip(char, cap_char):
    print(item_1, item_2)
```

```python
list(zip(char, cap_char))
```

```python
list(zip('Hello', 'World'))
```

```python
def odds(max_odd):
    out_list = list()

    for num in range(max_odd):
        if num %2 == 1:
            out_list.append(num)

    return out_list
```

```python
odds(13)
```

```python
def multiply_list(a, b):
    if len(a) != len(b):
        print('Lists must of of same length.')
        return None
    else:
        out = list()
        for item_1, item_2 in zip(a, b):
            out.append(item_1 * item_2)
        return out
```
