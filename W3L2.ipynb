### Loop pattern

```python
for index in range(10):
    print(index)
```

```python
char = ['a', 'b', 'c']
for index, item in enumerate(char):
    print(index, item)
```

```python
char = ['a', 'b', 'c']
cap_char = ['A', 'B', 'C']

for item_1, item_2 in zip(char, cap_char):
    print(item_1, item_2)
```

```python
list(zip(char, cap_char))
```

```python
list(zip('Hello', 'World'))
```
