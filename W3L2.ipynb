```python
for index in range(10):
    print(index)
```

```python
char = ['a', 'b', 'c']
for index, item in enumerate(char):
    print(index, item)
```

```python
char = ['a', 'b', 'c']
cap_char = ['A', 'B', 'C']

for item_1, item_2 in zip(char, cap_char):
    print(item_1, item_2)
```

```python
def odds(max_odd):
    out_list = list()

    for num in range(max_odd):
        if num %2 == 1:
            out_list.append(num)

    return out_list
```

```python
odds(13)
```

```python
def multiply(a, b):
    if isinstance(a, (float, int)) and isinstance(b, (float, int)):
        return a * b

    elif isinstance(a, list) and isinstance(b, list):
        return multiply_list(a, b)

    elif isinstance(a, list) and isinstance(b, (float, int)):
        return multiply_scalar_list(b, a)

    elif isinstance(a, (float, int)) and isinstance(b, list):
        return multiply_scalar_list(a, b)

    else:
        return None
        raise ValueError('Invalid inputs.')

def multiply_list(a, b):
    if len(a) != len(b):
        print('Lists must of of same length.')
        return None
    else:
        out = list()
        for item_1, item_2 in zip(a, b):
            out.append(item_1 * item_2)
        return out

def multiply_scalar_list(scalar, b):
    out = list()

    for item in b:
        out.append(scalar * item)
        return out
```

```python
print(multiply(2, 2))
print(multiply(2, [2, 5, 8]))
print(multiply([1, 2, 3], [4, 6, 2]))
```

```python
# Normal function definition

def square(x):
    return x * x

# Lambda definition

cube = lambda x: x * x * x
print(square(5), cube(5))
```
