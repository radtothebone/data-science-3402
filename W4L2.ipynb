```python
class Parent:
    def __init__(self):
        self.public = 'i am public'
        self._protected = 'i am protected'
        self.__private = 'i am private'

    def set_private(self, v):
        self.__private = v

    def get_private(self):
        return self.__private

class Child(Parent):
    def __init__(self, init_parent=True):
        if init_parent:
            super(Child, self).__init__()

    def set_public(self, v):
        self.public = v

    def set_protected(self, v):
        self._protected = v

    def get_public(self):
        return self.public

    def get_protected(self):
        return self._protected
```

```python
p = Parent()
c = Child()

# --- public: accessible anywhere ---
print(p.public)          # works
print(c.get_public())    # works
c.set_public('updated')
print(c.get_public())    # 'updated'

print(p._protected)      # works, but bad practice outside the class
print(c.get_protected()) # correct way
c.set_protected('updated protected')
print(c.get_protected()) # 'updated protected'

print(p.get_private())   # works via Parent's getter
c.set_private('updated private')
print(c.get_private())   # works because Child inherited Parent's getter/setter

# This would raise AttributeError — private is truly inaccessible from outside:
try:
    print(p.__private)
except AttributeError as e:
    print(f"AttributeError: {e}")

# Name mangling — this is how Python actually stores it:
print(p._Parent__private)  # works, but you should never do this in real code
```

```python
c_2 = Child()
c_2.set_private('Child is private')
print(c_2.get_private())
```
