```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
def factorial_zero(n):
    if n == 1:
        return 1
    else:
        return (n, factorial_zero(n-1))
```

```python
def factorial_one(n):
    prod = 1
    for e in range(1, n + 1):
        prod *= e
    return prod
```

```python
# Loops in recursion

def countdown(n):
    print(n)
    if n == 0:
        return
    else:
        countdown(n - 1)
```

```python
def rec_range(start, stop=None, step = 1):
    # Properly handle default arguments
    if stop:
        pass
    else:
        stop = start
        start = 0

    if start < stop:
        return [start] + rec_range(start + step, stop, step)
    else:
        return []
```

```python
rec_range(2, stop=20, step=2)
```

```python
def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return recur_fibo(n - 1) + recur_fibo(n - 2)
```

```python
recur_fibo(10)
```

```python
fibo_list = []

for i in range(10):
    fibo_list.append(recur_fibo(i))

fibo_list
```

```python
def leaf_count(n):
    # Initialize a counter to keep track of the total number of leaves.
    count = 0
    # Visit each item in the current list.
    for item in n:
        # If the item is itself a list, recursively count
        # the leaves contained within that sublist.
        if isinstance(item, list):
            count += leaf_count(item)
        # Otherwise, the item is a leaf node, so add 1.
        else:
            count += 1
    # Return the total number of leaves found in this list
    # and all of its nested sublists.
    return count
```

```python
# Checkers "AI" Player

# Scoring function
def score_board(board, player):
    return count_pieces(board, player) - count_pieces(board, switch_player(player))
```

### Considerations for the auto-play function:
1. The board
2. The player who is trying to win
3. The player whose turn we are currently considering
4. The depth of the tree

What do we return?
The move and the score

```python
import copy
def generate_moves(board, player, current_player, depth, keep_all_moves = False):
    if depth == 0:
        return list(), score_board(board, player)

    moves = []
    scores = []

    # Look through the board for pieces belonging to specific players
    for i in range(size):
        for j in range(size):
            if board[i][j] == current_player:
                for move in [left_move, right_move]:
                    new_board = copy.deepcopy(board)
                    if move_piece(new_board, current_player, (i,j), move, verbose=False):
                        next_moves, score = generate_moves(new_board, player, switch_player(current_player), depth - 1, keep_all_moves = keep_all_moves)
                        this_move = [(current_player, (i,j), move)]
                        if keep_all_moves:
                            this_move.extend(next_moves)
                            moves.append(this_move)
                            scores.append(score)
    return moves, scores
```
