```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
def factorial_zero(n):
    if n == 1:
        return 1
    else:
        return (n, factorial_zero(n-1))
```

```python
def factorial_one(n):
    prod = 1
    for e in range(1, n + 1):
        prod *= e
    return prod
```

```python
# Loops in recursion

def countdown(n):
    print(n)
    if n == 0:
        return
    else:
        countdown(n - 1)
```

```python
countdown(5)
```

```python
def rec_range(start, stop=None, step = 1):
    # Properly handle default arguments
    if stop:
        pass
    else:
        stop = start
        start = 0

    if start < stop:
        return [start] + rec_range(start + step, stop, step)
    else:
        return []
```

```python
rec_range(2, stop=20, step=2)
```
