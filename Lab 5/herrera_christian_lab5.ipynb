## Problem 1
Factorial Calculation – Write a recursive function to compute the factorial of a given number `n`.

**Task:** Implement the recursive function below.

```python
def factorial(n):
	if n > 0:
		return n * factorial(n-1)
	else:
		return 1

factorial(5)
```

## Problem 2
Sum of Digits – Write a recursive function that takes an integer `n` and returns the sum of its digits.

**Task:** Implement the recursive function below.

```python
def sum_list(num):
	if num > 0:
		return num % 10 + sum_list(num // 10)
	else:
		return 0

sum_list(222)
```

## Problem 3
Reverse a String – Write a recursive function to reverse a string without using slicing.

**Task:** Implement the recursive function below.

```python
def reverse_string(s, i=None):
	if i is None:
		i = len(s) - 1
	if i == len(s):
		return ""
	if i < 0:
		return ""
	else:
		return s[i] + reverse_string(s, i - 1)

reverse_string("hello")
```

## Problem 4
Power Function – Write a recursive function that computes `x^n` (x to the power of n).

**Task:** Implement the recursive function below.

```python
def power(x, n):
	if n == 0:
		return 1
	else:
		return x * power(x, n - 1)

power(2, 5)
```

## Problem 5
Binary Search – Implement recursive binary search on a sorted list. Return the index if found, otherwise `-1`.

**Task:** Implement the recursive function below.

```python
def binary_search(arr, target, low=0, high=None):
	if high is None:
		high = len(arr) - 1
	if low > high:
		return -1
	mid = (low + high) // 2
	if arr[mid] == target:
		return mid
	elif arr[mid] < target:
		return binary_search(arr, target, mid + 1, high)
	else:
		return binary_search(arr, target, low, mid - 1)

numbers = [1, 3, 5, 7, 9, 11, 13]
print(binary_search(numbers, 7))
print(binary_search(numbers, 4))
```

## Problem 6
Generate All Subsets (Power Set) – Write a recursive function that returns all subsets of a list. Example: [1,2] → [[  ], [1], [2], [1,2]]

**Task:** Implement the recursive function below.

```python
def power_set(lst):
	if len(lst) == 0:
		return [[]]
	else:
		first = lst[0]
		rest = lst[1:]
		rest_subsets = power_set(rest)
		all_subsets = []
		for subset in rest_subsets:
			all_subsets.append(subset)
		for subset in rest_subsets:
			all_subsets.append([first] + subset)
		return all_subsets

power_set([1, 2])
```

## Problem 7
Sum of List – Compute the sum of all elements in a list recursively.

**Task:** Implement the recursive function below.

```python
def sum_list_elements(lst):
	if len(lst) == 0:
		return 0
	else:
		return lst[0] + sum_list_elements(lst[1:])

sum_list_elements([1, 2, 3, 4])
```

## Problem 8
Find Minimum in List – Find the minimum element in a list recursively (without using `min()`).

**Task:** Implement the recursive function below.

```python
def find_min(lst):
	if len(lst) == 0:
		return None
	if len(lst) == 1:
		return lst[0]
	else:
		rest_min = find_min(lst[1:])
		if lst[0] < rest_min:
			return lst[0]
		else:
			return rest_min

find_min([5, 3, 8, 1, 9])
```

## Problem 9
Decimal to Binary Conversion – Convert a decimal number into binary using recursion.

**Task:** Implement the recursive function below.

```python
def decimal_to_binary(n):
	if n < 2:
		return str(n)
		n = n - 1
	else:
		return decimal_to_binary(n // 2) + str(n % 2)

decimal_to_binary(10)
```

## Problem 10
String Length – Calculate the length of a string recursively (without using `len()`).

**Task:** Implement the recursive function below.

```python
def string_length(s):
	if s == "":
		return 0
	else:
		return 1 + string_length(s[1:])

string_length("hello")
```

## Problem 11
Nested List Sum – Sum all integers inside a nested list recursively. Example: [1, [2, [3, 4]], 5] → 15.

**Task:** Implement the recursive function below.

```python
def nested_sum(lst):
	total = 0
	count = 0
	for item in lst:
		if isinstance(item, list):
			total = total + nested_sum(item)
		else:
			total = total + item
			count = count + 1
	return total

nested_sum([1, [2, [3, 4]], 5])
```

## Problem 12
Permutations of String – Print all permutations of a string using recursion. Example: "abc" → ['abc', 'acb', 'bac', 'bca', 'cab', 'cba'].

**Task:** Implement the recursive function below.

```python
def permutations(s):
	if s == "":
		return [""]
	if len(s) <= 1:
		return [s]
	else:
		result = []
		for i in range(len(s)):
			char = s[i]
			remaining = s[:i] + s[i+1:]
			for perm in permutations(remaining):
				result.append(char + perm)
		return result

permutations("abc")
```
