### Classes

```python
class student:
    def __init__(self, name, id_number, year):
        self.name = name
        self.id_number = id_number
        self.year = year
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

    def print_grades(self):
        for grade in self.grades:
            print(grade)
```

```python
a_student = student('Johnny Doe', 1111, 2026)

a_student.add_grade(99)
a_student.add_grade(90)
a_student.add_grade(87)

a_student.print_grades()
```

```python
print('Average:', a_student.average_grade())
```

### Subclasses

```python
class Person:
    def __init__(self, name, id_number):
        self.name = name
        self.id_number = id_number

class Student(Person):
    def __init__(self, name, id_number, year):
        super().__init__(name, id_number)
        self.year = year
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

    def print_grades(self):
        for grade in self.grades:
            print(grade)


class Faculty(Person):
    def __init__(self, name, id_number):
        super().__init__(name, id_number)
        self.courses = []

    def add_courses(self, course):
        self.courses.append(course)

    def print_courses(self):
        for course in self.courses:
            print(course)
```

```python
class Device:
    def __init__(self):
        self.__update()

    def operate(self):
        print('operate')

    def __update(self): # Private method
        print('Updating software')
```

```python
a_device = Device()
a_device.operate()
```

```python
class Car:
    def __init__(self, name='Unnamed', n_doors=4, max_passengers=4):
        self.__n_passengers = 0
        self.__name = name
        self.__n_doors = n_doors
        self.__max_passengers = max_passengers

    # Getters
    def get_name(self):
        return self.__name

    def get_n_passengers(self):
        return self.__n_passengers

    def get_n_doors(self):
        return self.__n_doors

    def get_max_passengers(self):
        return self.__max_passengers

    # Setter
    def set_name(self, name):
        if isinstance(name, str):
            self.__name = name
        else:
            print('Name must be a string. ')

    def add_passenger(self, n=1):
        if isinstance(n, (int, float)):
            self.__n_passengers += n
            if self.__n_passengers > self.__max_passengers:
                left_outside = self.__n_passengers - self.__max_passengers
                self.__n_passengers = self.__max_passengers
                print('car is full,', left_outside, 'were left outside')
        else:
            print('number of passengers must be an integer')

    def remove_passenger(self, n=1):
        if isinstance(n, (int, float)):
            self.__n_passengers -= n
            if self.__n_passengers < 0:
                self.__n_passengers = 0
        else:
            print('Number of passengers to remove must be an integer. ')
```

```python
my_car = Car()
print(my_car.name())
```
