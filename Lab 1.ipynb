```python
player_1 = 'B'
player_2 = 'Q'
empty = 0

size = 8
```

```python
def make_game_board(board_size: int):
    board = [[empty] * board_size for _ in range(board_size)]

    for i in range(0, board_size):
        board[0][i] = player_1
        board[-1][i] = player_2

    for i in range(0, board_size):
        board[0][i] = player_1
        board[1][i] = player_1
        board[-2][i] = player_2

    # Adding the pawn pieces across the rows
    for i in range(0, board_size):
        board[1][i] = board[1][i] + 'P'
        board[6][i] = board[6][i] + 'P'

    # Adding the Rook pieces
    r_coords = [[0,0], [0,7], [7,7], [7,0]]
    for coord in r_coords:
        board[coord[0]][coord[1]] = board[coord[0]][coord[1]] + 'R'

    # Adding the Knight pieces
    n_coords = [[0,1], [0,6], [7,6], [7,1]]
    for coord in n_coords:
        board[coord[0]][coord[1]] = board[coord[0]][coord[1]] + 'N'

    # Adding the Bishop pieces
    b_coords = [[0,2], [0,5], [7,5], [7,2]]
    for coord in b_coords:
        board[coord[0]][coord[1]] = board[coord[0]][coord[1]] + 'B'

    k_coords = [[0,4], [7,4]]
    for coord in k_coords:
        board[coord[0]][coord[1]] = board[coord[0]][coord[1]] + 'K'

    q_coords = [[0,3], [7,3]]
    for coord in q_coords:
        board[coord[0]][coord[1]] = board[coord[0]][coord[1]] + 'Q'

    return board

make_game_board(8)
```
